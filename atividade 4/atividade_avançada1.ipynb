```python
import numpy as np
import matplotlib.pyplot as plt

# Carregar os dados
dados = np.load('todas_posicoes_passeio.npz')

# Extrair os valores de posição
posicoes_05 = dados['posicoes_p05']
posicoes_06 = dados['posicoes_p06']

# "Achatar" os arrays para transformá-los em 1D
posicoes_05_flat = posicoes_05.flatten()
posicoes_06_flat = posicoes_06.flatten()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(posicoes_05_flat, bins=30, alpha=0.7, color='green')
plt.xlabel('Posição')
plt.ylabel('Frequência')
plt.title('Histograma das posições (p = 0.5)')

plt.subplot(1, 2, 2)
plt.hist(posicoes_06_flat, bins=30, alpha=0.7, color='blue')
plt.xlabel('Posição')
plt.ylabel('Frequência')
plt.title('Histograma das posições (p = 0.6)')

plt.tight_layout()
plt.show()
```
