```python
import numpy as np
import matplotlib.pyplot as plt

# Carregar os dados
dados1 = np.load('Dados_passeio_p_0.5.npz')
dados2 = np.load('Dados_passeio_p_0.8.npz')

# Extrair os valores de posição
posicoes1 = dados1['valores_posicao']
posicoes2 = dados2['valores_posicao']

plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
plt.hist(posicoes1, bins=30, alpha=0.7, color='green')
plt.xlabel('Posição')
plt.ylabel('Frequência')
plt.title('Histograma das posições (p = 0.5)')


plt.subplot(1, 2, 2)
plt.hist(posicoes2, bins=30, alpha=0.7, color='blue')
plt.xlabel('Posição')
plt.ylabel('Frequência')
plt.title('Histograma das posições (p = 0.8)')

plt.tight_layout()
plt.show()
```
