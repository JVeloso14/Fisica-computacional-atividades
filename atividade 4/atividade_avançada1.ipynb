```python
import numpy as np
import matplotlib.pyplot as plt


dados = np.load('todas_posicoes_passeio.npz')


posicoes_05 = dados['posicoes_p05']  #(100 simulações, número de passos)
posicoes_06 = dados['posicoes_p06']


todas_posicoes_05 = []
todas_posicoes_06 = []

# Para cada simulação, coleta todas as posições
for sim in range(posicoes_05.shape[0]):
    todas_posicoes_05.extend(posicoes_05[sim])
    todas_posicoes_06.extend(posicoes_06[sim])

plt.figure(figsize=(12, 5))

# Histograma para p=0.5
plt.subplot(1, 2, 1)
plt.hist(todas_posicoes_05, bins=30, alpha=0.7, color='green')
plt.xlabel('Posição')
plt.ylabel('Frequência')
plt.title('Histograma das posições - 100 simulações (p = 0.5)')

# Histograma para p=0.6
plt.subplot(1, 2, 2)
plt.hist(todas_posicoes_06, bins=30, alpha=0.7, color='blue')
plt.xlabel('Posição')
plt.ylabel('Frequência')
plt.title('Histograma das posições - 100 simulações (p = 0.6)')

plt.tight_layout()
plt.show()
```
