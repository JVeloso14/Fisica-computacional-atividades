```python
import numpy as np
import matplotlib.pyplot as plt


dados1 = np.load('Dados_passeio_p_0.5.npz')
dados2 = np.load('Dados_passeio_p_0.8.npz')

posicoes1 = dados1['valores_posicao']
posicoes2 = dados2['valores_posicao']


plt.figure(figsize=(12, 5))

# Função para extrair fatias dos dados: início, meio e fim
def extrair_momentos(posicoes):
    n = len(posicoes)
    inicio = posicoes[:n//3]
    meio   = posicoes[n//3:2*n//3]
    fim    = posicoes[2*n//3:]
    return inicio, meio, fim

# Histogramas para p = 0.5
plt.subplot(1, 2, 1)
inicio1, meio1, fim1 = extrair_momentos(posicoes1)
plt.hist(inicio1, bins=30, alpha=0.7, color='blue', label='Início')
plt.hist(meio1,   bins=30, alpha=0.7, color='green', label='Meio')
plt.hist(fim1,    bins=30, alpha=0.7, color='red', label='Fim')
plt.title('Histograma (p = 0.5)')
plt.xlabel('Posição')
plt.ylabel('Frequência')
plt.legend()

# Histogramas para p = 0.8
plt.subplot(1, 2, 2)
inicio2, meio2, fim2 = extrair_momentos(posicoes2)
plt.hist(inicio2, bins=30, alpha=0.7, color='blue', label='Início')
plt.hist(meio2,   bins=30, alpha=0.7, color='green', label='Meio')
plt.hist(fim2,    bins=30, alpha=0.7, color='red', label='Fim')
plt.title('Histograma (p = 0.8)')
plt.xlabel('Posição')
plt.ylabel('Frequência')
plt.legend()

plt.tight_layout()
plt.show()
```
