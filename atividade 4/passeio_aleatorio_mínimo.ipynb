# Passeio aleatório mínimo

Objetivo: Criar um gráfico da posição do "bebado" em função do número de passos. 

Ele tem uma probilidade p de ir para a direita e 1-p para ir a esquerda (cada passo equivale a uma unidade). 

```python
import numpy as np
import matplotlib.pyplot as plt 

class PasseioAleatorio:
    def __init__(self, p, N):
        """

        Parâmetros:
        p (float): Probabilidade de movimento para a direita
        N (int): Número de passos no passeio
        """
        self.p = p
        self.N = N
    
    def função_passeio(self):
        """
        Gera um passeio aleatório usando rng.choice() e np.cumsum()
        
        Retorna:
        array: Posições do passeio aleatório
        """
        rng = np.random.default_rng()
        passos = rng.choice([1, -1], size=self.N, p=[self.p, 1-self.p])
        
        x = np.concatenate(([0], np.cumsum(passos)))
        
        return x


p = [0.5, 0.8]        

for i in p:
    valores = PasseioAleatorio(i, 1000)
    valores_posicao = valores.função_passeio()
    caminho_arquivo = f'C:/Users/joaof/OneDrive/Área de Trabalho/Fisica computacionatividades/atividade 4/Dados_passeio_p_{i}'
    np.savez(caminho_arquivo, valores_posicao= valores_posicao, p=i )
    plt.plot(range(len(valores_posicao)), valores_posicao , label = 'posicao')
    plt.ylabel('x(N)')
    plt.xlabel('N')
    plt.legend()
    plt.title(f'O andar do bebado com p={i}')
    plt.show()
    

```
