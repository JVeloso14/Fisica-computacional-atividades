```python
import matplotlib.pyplot as plt
import numpy as np 

x= np.linspace(-np.pi , np.pi , 2000)

def Teste_prova(N):
    soma = 0
    for i in range(1,N-1):
        soma += np.pi/2 +  (1-(-1)**i)/(2*np.pi * i**2) * np.cos(i*x)
    return soma

soma  = Teste_prova(10000)

plt.plot(x, soma)
plt.show()
```
