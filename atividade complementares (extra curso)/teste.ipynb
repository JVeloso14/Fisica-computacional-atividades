```python
import numpy as np
import matplotlib.pyplot as plt


k_B = 1.0
N = 1.0
k = 1.0


T_list = [0, 1, 2, 3]
H_vals = np.linspace(-3, 3, 300)

# Função para a Energia Livre de Helmholtz F(T,H)
# De acordo com a expressão fornecida:
#   F = - k_B * T * N * [ k + ln( cosh(H) + sqrt( sinh^2(H) + e^(-4k) ) ) ]
def F_helmholtz(H, T, N, k):
    if T == 0:
        # Por exemplo, retornar F no limite T->0 (caso queira uma análise especial).
        # Abaixo é só um placeholder que evita erro de log(0).
        return np.zeros_like(H) * np.nan
    
    return -k_B * T * N * ( 
        k + np.log( 
            np.cosh(H) + np.sqrt( np.sinh(H)**2 + np.exp(-4*k) )
        )
    )

# Cria a figura
plt.figure(figsize=(6, 4))


for i, T in enumerate(T_list):
    F_vals = F_helmholtz(H_vals, T, N, k)
    if i == 0:
        linestyle = '-'
    elif i == 1:
        linestyle = '--'
    elif i == 2:
        linestyle = '-.'
    else:
        linestyle = ':'
    
    plt.plot(H_vals, F_vals, linestyle, label=f"$k_B T = {T}$")

plt.xlabel("$H$")
plt.ylabel("$F(T,H)$")
plt.title("Energia Livre de Helmholtz vs. H")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
