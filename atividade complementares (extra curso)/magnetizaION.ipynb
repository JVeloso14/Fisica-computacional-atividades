```python
import numpy as np
import matplotlib.pyplot as plt

def magnetization(H, T):
    """
    Retorna m(T,H) = sinh(H) / sqrt( sinh^2(H) + e^(-4K) ),
    onde K = 1/T se J/k_B=1.
    """
    if T == 0:
        m = np.sign(H)
        m[H==0] = 0
        return m
    else:
       
        K = 1.0 / T
        num = np.sinh(H)
        den = np.sqrt(np.sinh(H)**2 + np.exp(-4.0*K))
        return num / den


T_list = [0, 1, 2, 3]


H_vals = np.linspace(-3, 3, 400)


linestyles = ['-', '--', '-.', ':']

plt.figure(figsize=(6, 4))

for i, T in enumerate(T_list):
    m_vals = magnetization(H_vals, T)
    plt.plot(H_vals, m_vals, linestyles[i], label=f"$T = {T}$")

plt.axhline(0, color='k', linewidth=0.8)  # linha horizontal no 0 (opcional)
plt.xlabel("$H$")
plt.ylabel("$m(T,H)$")
plt.title("Magnetização vs. H ")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
