```python
import matplotlib.pyplot as plt
import numpy as np

# Função para calcular a taxa de variação de v
def f(t, v, b, g):
    return -b * v**2 + g

# Método de Runge-Kutta de quarta ordem (RK4)
def RK4(v0, b, g, t0, tmax, dt):
    t = t0
    v = v0
    tempo = [t0]
    velocidade = [v0]

    while t < tmax:
        k1 = f(t, v, b, g)
        k2 = f(t + 0.5 * dt, v + 0.5 * dt * k1, b, g)
        k3 = f(t + 0.5 * dt, v + 0.5 * dt * k2, b, g)
        k4 = f(t + dt, v + dt * k3, b, g)

        v = v + (dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
        t = t + dt

        tempo.append(t)
        velocidade.append(v)

    return tempo, velocidade

# Parâmetros
v0 = 0  
b = 0.035 
g = 10 
t0 = 0 
tmax = 10
dt = 0.01 


velocidade_terminal = np.sqrt(g / b)


tempo, velocidade = RK4(v0, b, g, t0, tmax, dt)

vel_terminal_line = [velocidade_terminal] * len(tempo)

# Plotando os resultados
plt.figure(figsize=(9, 6))
plt.plot(tempo, velocidade, label='Velocidade')
plt.plot(tempo, vel_terminal_line, label=f'Velocidade terminal = {velocidade_terminal:.2f} m/s', color='red', linestyle='--')
plt.title('Velocidade vs. Tempo', fontsize=14)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (m/s)')
plt.grid(False)
plt.legend()
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.show()
```
