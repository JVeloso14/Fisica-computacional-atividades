```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp

t = np.linspace(0, 15, 200)

v0= [0] 
g = 10 #m/s^2
k= 3.5
m= 100

#Eq diferencial da queda com resistencia do ar: v' = g - (k/m )v^2

def queda_resis(t , v):
    return g - (k/m)*v**2

solucao = solve_ivp(queda_resis , [0 , 15], v0 , t_eval = t, method = 'RK45')

plt.plot(solucao.t , solucao.y[0], label = 'v(t)' )
plt.xlabel('t (s)')
plt.ylabel('v(t)')

plt.show()
```
