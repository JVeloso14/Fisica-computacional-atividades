```python
#Dente de serra utilizando Fourier 

import matplotlib.pyplot as plt
import numpy as np 

x= np.linspace(-3*np.pi , 3*np.pi , 2000)

def Fourier_dente_serra(N):
    soma = 0
    for i in range(1,N-1):
        soma += 2*((-1)**(i+1)*np.sin(i*x))/i
    return soma

soma  = Fourier_dente_serra(10000)

plt.plot(x, soma)
plt.show()
```
