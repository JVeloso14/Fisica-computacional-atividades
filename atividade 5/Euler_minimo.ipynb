```python
import matplotlib.pyplot as plt
import numpy as np


b = 0.1  #kg/m
g= 9.8 #m/s^2
m = 1 #kg
v0 =0 #velocidade inicial
h= 0.1 #passos 

def Queda_resistencia(v):
    return -g -b/m *v

t= np.arange(0, 100 , h)
v = np.zeros(len(t))

v[0] = v0

for i in range(1, len(t)):
    v[i] = v[i-1] + h*Queda_resistencia(v[i-1])
    
for i in range(len(t) -1):
    print(f't = {t[i]} , v = {v[i]}')

plt.plot(t , v)
plt.xlabel('t(s)')
plt.ylabel('v (m/s)')
plt.tight_layout()
plt.title('Queda livre com resistencia do ar')
plt.show()

```
