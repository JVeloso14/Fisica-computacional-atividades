```python

import numpy as np
import matplotlib.pyplot as plt


N = 100
L = 1.0  
dx = L / (N - 1) 

# Construir a matriz D
diagonal = -2 * np.ones(N-2)
diagonal_2 = np.ones(N-3)
D = (1 / dx**2) * (np.diag(diagonal) + np.diag(diagonal_2, 1) + np.diag(diagonal_2, -1))

# Calcular autovalores e autovetores
autovalores, autovetores = np.linalg.eigh(D)
autovalor_inv = autovalores[::-1] 
autovetor_inv = autovetores[:, ::-1] 
n = np.arange(1, N-1)

for modo in range(5):
    plt.plot(autovetor_inv[:, modo], label=f'Modo {modo+1} (λ = {autovalor_inv[modo]:.2f})')
    plt.title('Modos normais')
    plt.legend()
    plt.show()


```
