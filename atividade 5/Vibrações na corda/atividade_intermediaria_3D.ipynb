```python

import numpy as np
import matplotlib.pyplot as plt


N = 100
L = 1.0  
dx = L / (N - 1) 

# Construindo a matriz D
diagonal = -2 * np.ones(N-2)
diagonal_2 = np.ones(N-3)
D = (-1 / dx**2) * (np.diag(diagonal) + np.diag(diagonal_2, 1) + np.diag(diagonal_2, -1))

# Calcula os autovalores e autovetores
autovalores, autovetores = np.linalg.eigh(D)
n = np.arange(1, N)

for modo in range(4):

    psi = autovetores[:, modo]
    psi = np.concatenate(([0], psi, [0]))  # Adiciona as condições de contorno para "fechar" a envoltória
    plt.plot(psi, color = 'blue', label=f'Modo {modo+1} (λ = {autovalores[modo]:.2f})')
    plt.plot(-psi, color='blue')
    plt.title(f'{modo+1}º Modo de vibração')
    plt.legend()
    plt.xlabel('Pontos na malha')
    plt.ylabel('Autovetores')
    plt.show()


```
