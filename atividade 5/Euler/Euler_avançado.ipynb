```python
import matplotlib.pyplot as plt
import numpy as np


def Queda_Resistencia(t, v, b, g):
    return -g - (b/m) * v 


def RK4(v0, b, g, t0, tmax, dt):
    t = t0
    v = v0
    tempo = [t0]
    velocidade = [v0]

    while t < tmax:
        k1 = Queda_Resistencia(t, v, b, g)
        k2 = Queda_Resistencia(t + 0.5 * dt, v + 0.5 * dt * k1, b, g)
        k3 = Queda_Resistencia(t + 0.5 * dt, v + 0.5 * dt * k2, b, g)
        k4 = Queda_Resistencia(t + dt, v + dt * k3, b, g)

        v = v + (dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
        t = t + dt
        tempo.append(t)
        velocidade.append(v)
    return tempo, velocidade

print(velocidade)

# Parmetros
v0 = 0  
b = 0.1 
g = 9.8
t0 = 0 
m=1
tmax = 100
dt = 0.1 


velocidade_terminal = (-m*g )/ b


tempo, velocidade = RK4(v0, b, g, t0, tmax, dt)

vel_terminal_line = [velocidade_terminal] * len(tempo)

# Plotando os resultados
plt.figure(figsize=(9, 6))
plt.plot(tempo, velocidade, label='Velocidade')
plt.plot(tempo, vel_terminal_line, label=f'Velocidade terminal = {velocidade_terminal:.2f} m/s', color='red', linestyle='--')
plt.title('Velocidade vs. Tempo', fontsize=14)
plt.xlabel('Tempo (s)')
plt.ylabel('Velocidade (m/s)')
plt.grid(False)
plt.legend()
plt.show()
```
