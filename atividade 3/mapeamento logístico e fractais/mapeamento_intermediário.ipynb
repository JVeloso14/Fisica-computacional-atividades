```python
import numpy as np
import matplotlib.pyplot as plt

class MapaLogistico():
    def __init__(self, r, x0, N):
        self.r = r
        self.x0 = x0
        self.N = N
        
    def mapa_logistico(self):
        x_n = [self.x0]
        for i in range(self.N - 1):
            novo_valor = self.r * x_n[i] * (1 - x_n[i])
            x_n.append(novo_valor)
        return x_n

    
    

M = 1000
N = 1000 + M
lista_r = np.arange(1, 4, 0.1)

plt.figure(figsize=(8, 10))

for r in lista_r:
    simulacao = MapaLogistico(r, 0.5, N).mapa_logistico()
    plt.plot([r]*np.ones(M), simulacao[-M:], 'k.', markersize=1)

plt.xlabel('r')
plt.ylabel('x_n')
plt.tight_layout()
plt.show()
```
