# Mapeamento Logístico- atividade mínima

O mapeamento logístico é um modelo matemático simples que descreve como uma população evolui ao longo do tempo, considerando fatores de crescimento e limitação. Ele é definido pela equação:

$$
x_{n+1} = r \cdot x_n \cdot (1 - x_n)
$$

Onde:
- $x_n$ representa a população normalizada no instante \(n\) (valores entre 0 e 1).
- $r$ é a taxa de crescimento populacional.
- $x_{n+1}$ é o valor da população no próximo instante.

No código fornecido:
1. A classe `MapaLogistico` é criada para encapsular o comportamento do mapeamento logístico.
    - O construtor (`__init__`) recebe os parâmetros:
      - `r`: taxa de crescimento.
      - `x0`: valor inicial da população.
      - `N`: número de iterações.
    - O método `mapa_logistico` calcula a evolução da população ao longo de $N$ iterações, armazenando os valores em uma lista.

2. O código utiliza diferentes valores de $r$ (`r_values = [2, 3.3, 3.5, 3.9]`) para simular o comportamento do sistema:
    - Para $r < 3$, o sistema tende a um ponto fixo.
    - Para $3 \leq r < 3.57$, o sistema apresenta comportamento periódico.
    - Para $r \geq 3.57$, o sistema entra em caos, exibindo comportamento imprevisível.



```python
import numpy as np
import matplotlib.pyplot as plt

class MapaLogistico():
    def __init__(self, r, x0, N):
        self.r = r
        self.x0 = x0
        self.N = N
        
    def mapa_logistico(self):
        x_n = [self.x0]
        for i in range(self.N - 1):
            novo_valor = self.r * x_n[i] * (1 - x_n[i])
            x_n.append(novo_valor)
        return x_n

r_values = [2, 3.3, 3.5, 3.9]

n = list(range(1, 1001))

plt.figure(figsize=(8, 10))

for i, r in enumerate(r_values):
    resultado = MapaLogistico(r, 0.5, 1000)
    simulacao = resultado.mapa_logistico()
    plt.subplot(len(r_values), 1, i + 1)
    plt.plot(n, simulacao, '.', markersize=1)
    plt.title(f'r = {r}')
    plt.ylabel('x_n')

plt.xlabel('n')
plt.tight_layout()
plt.show()
```
