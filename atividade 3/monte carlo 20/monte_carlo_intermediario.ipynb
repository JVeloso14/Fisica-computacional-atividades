# Simulação de Monte Carlo para calcular o valor de $\pi$-  Intermediário



```python

import numpy as np
from numpy import random


def calc_pi(pontos,retorna_pontos=False):
    pontos_internos = []
    pontos_externos = []
    assert isinstance(retorna_pontos, bool) 
    assert isinstance(pontos, int) 

    for i in range(pontos):
        x= random.uniform(-1,1)
        y= random.uniform(-1,1)
        if x**2 + y**2 <=1:
            pontos_internos.append((x,y))
        else:
            pontos_externos.append((x,y))
        estimativa_pi  = 4 * len(pontos_internos)/ pontos
    
    if retorna_pontos == True:
        return estimativa_pi, pontos_internos
    else:
        return estimativa_pi

calc_pi(10000, True )
```
