# Bhaskara - atividade 2 mínima

Neste código, os três casos da equação quadrática ($\Delta < 0$, $\Delta = 0$, $\Delta > 0$) foram abordados a partir das constantes \(a, b, c, d, e, f, g, h, i\). A função ```Bhaskara``` é a mesma utilizada na atividade 1, porém sem a verificação ```assert``` para $\Delta < 0$. As raízes correspondentes a essas constantes foram encontradas definindo ```raizes```, ```raizes2``` e ```raizes3```.  

Para o plot de cada gráfico, foi utilizada a biblioteca ```matplotlib.pyplot``` por meio do comando ```plt.subplot()```, que possui três parâmetros: o primeiro corresponde ao número de **linhas** da grade de subgráficos, o segundo ao número de **colunas**, e o terceiro indica a **posição** do subplot dentro da grade.  

Para identificar as raízes no gráfico, foi utilizado o loop ```for raiz in raizes``` para percorrer cada solução em ```raizes = Bhaskara(a, b, c)```,  
bem como o comando ```plt.scatter()``` para marcar apenas as raízes.  


````python
import numpy as np
import matplotlib.pyplot as plt

def Bhaskara(a, b, c):
    """
    Calcula as raízes reais da equação quadrática ax² + bx + c = 0 utilizando a fórmula de Bhaskara.

    Parâmetros:
        a (float): Coeficiente quadrático (não pode ser zero).
        b (float): Coeficiente linear.
        c (float): Termo constante.

    Retorna:
        tuple: Contém uma ou duas raízes reais da equação. Se o discriminante for negativo, retorna uma tupla vazia.
        
    """
    delta = b**2 - 4 * a * c
    assert a != 0, "O coeficiente 'a' não pode ser zero"

    if delta < 0:
        return ()  
    elif delta == 0:
        x = -b / (2 * a)
        return (x,)
    else:
        x1 = (-b + (delta)**(1/2)) / (2 * a)
        x2 = (-b - (delta)**(1/2)) / (2 * a)
        return (x1, x2)

# Definição dos coeficientes das equações quadráticas
a, b, c = 1, -5, 6
d, e, f = 1, -10, 25
g, h, i = 1, 2, 3

# Cálculo das raízes para cada conjunto de coeficientes
raizes = Bhaskara(a, b, c)
raizes2 = Bhaskara(d, e, f)
raizes3 = Bhaskara(g, h, i)

# Definição do intervalo de x para os gráficos
x = np.linspace(-2, 6, 200)

# Cálculo dos valores das funções quadráticas
y = a * x**2 + b * x + c
y1 = d * x**2 + e * x + f
y2 = g * x**2 + h * x + i

# Criando os subplots para visualizar as equações quadráticas
plt.figure(figsize=(8, 18))

# Primeiro gráfico (duas raízes reais)
plt.subplot(3, 1, 1)
plt.plot(x, y, label=f'f(x) = {a}x² + {b}x + {c}', color='blue')
for raiz in raizes:
    plt.scatter(raiz, 0, color='red', zorder=5, label=f'Raiz: x = {raiz:.2f}')
plt.title("Equação quadrática com duas raízes reais")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Segundo gráfico (uma raiz real)
plt.subplot(3, 1, 2)
plt.plot(x, y1, label=f'f(x) = {d}x² + {e}x + {f}', color='green')
for raiz in raizes2:
    plt.scatter(raiz, 0, color='blue', zorder=5, label=f'Raiz: x = {raiz:.2f}')
plt.title("Equação quadrática com uma raiz real")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Terceiro gráfico (sem raízes reais)
plt.subplot(3, 1, 3)
plt.plot(x, y2, label=f'f(x) = {g}x² + {h}x + {i}', color='orange')
plt.title("Equação quadrática sem raízes reais")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````
