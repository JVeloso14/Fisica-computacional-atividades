```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes da equação quadrática
a = 1
b = -10
c = 25

# Função para calcular as raízes usando Bhaskara
def Bhaskara(a, b, c):
    delta = b**2 - 4 * a * c
    assert a != 0, "O coeficiente 'a' não pode ser zero"

    if delta < 0:
        raise ValueError('O delta não pode ser negativo')

    if delta == 0:
        x = -b / (2 * a)
        return (x,)
    else:
        x1 = (-b + (delta)**(1/2)) / (2 * a)
        x2 = (-b - (delta)**(1/2)) / (2 * a)
        return (x1, x2)


raizes = Bhaskara(a, b, c)


x = np.linspace(2, 8, 200)
y = a * x**2 + b * x + c


plt.figure(figsize=(8, 6))
plt.title("Gráfico de y = ax² + bx + c")
plt.plot(x, y, label=f'f(x) = {a}x² + {b}x + {c}', color='blue')

# Destacando as raízes no gráfico
for raiz in raizes:
    plt.scatter(raiz, 0, color='red', zorder=5, label=f'Raiz: x = {raiz:.0f}')  # Marcador das raízes


plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.ylim(-1, 6)  # Limitando o eixo y para melhor visualização
plt.xlim(2, 8)  # Limitando o eixo x para mostrar o comportamento negativo

# Exibindo o gráfico
plt.show()
```
