# Função Gaussiana - Atividade 2 Mínima  

A função Gaussiana, amplamente utilizada em muitas áreas da física, como a distribuição Gaussiana e a fórmula de Rodrigues para o oscilador harmônico quântico, é definida por  

$$f(x) = ae^{\frac{-(x-b)^2}{2c^2}},$$  

onde $a$ é a altura do pico da curva, $b$ é a posição do centro do pico e a constante $c$ está relacionada à espessura (como pode ser visualizado na função $\exp(-2x^2)$).  

Alterando os parâmetros $a$, $b$ e $c$, obtêm-se curvas com diferentes alturas, centros e espessuras. Quanto maior for $a$, maior será a altura do pico. Para valores positivos de $b$, a curva se desloca para a direita, ou seja, a posição do centro muda. À medida que $c$ diminui, a curva se torna mais estreita.

As funções utilizadas foram as mesmas abordadas nas atividades de Bhaskara e das funções exponenciais.


```python
import numpy as np
import matplotlib.pyplot as plt


x= np.linspace(-10 , 10 ,200)#definindo o x e o intervalo de -10 a 10 , com 200 subnintervalos

func1= np.exp(-x**2)
func2 = 1.5*np.exp(-x**2)
func3 = np.exp(-2*x**2)
func4 = np.exp(-(x-1)**2)
#As funções que serão plotadas y(x) vs x

plt.figure(figsize=(8, 6))
plt.title("Gráfico de y = exp(-x^2)")
plt.plot(x, func1, label='f(x) = exp(-x^2)', color='blue')
plt.plot(x, func2, label='f(x) = 1.5exp(-x^2)', color='red')
plt.plot(x, func3, label='f(x) = exp(-2x^2)', color='black')
plt.plot(x, func4, label='f(x) = exp(-(x-1)^2)', color='green')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.ylim(-0.5, 1.5)  # Limitando o eixo y para melhor visualização
plt.xlim(-10,10)    # Limitando o eixo x para mostrar o comportamento negativo
plt.show()
```
