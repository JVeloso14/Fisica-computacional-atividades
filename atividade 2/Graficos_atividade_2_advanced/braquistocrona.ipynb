```python
import numpy as np 
import matplotlib.pyplot as plt

theta = np.linspace(0 , 5 * np.pi , 100) #definindo o theta e o intervalo de 0 a 5pi , com 100 subnintervalos

x = 5*(theta - np.sin(theta))
y=  5*(1 - np.cos(theta))
#As funcoes x e y em funcao de theta

plt.plot(x,y , color = 'Blue')
plt.title('Braquistocrona')
plt.show()
```
