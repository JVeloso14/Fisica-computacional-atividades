# Deltoid - atividade 2 avançada

As coordenadas $x$ e $y$ em função de $\theta$ que descrevem o Deltoid são dadas por $$x = 2\cos{\theta} + \cos{2\theta},$$ $$y= 2\sin{\theta} - \sin{2\theta}.$$ Dessa forma, foi utilizado um intervalo de $\theta$ conveniente para a visualização da imagem (de 0 a $5\pi$).

Obs: as funções do ```numpy``` e do ```matplotlib.pyplot``` para gerar o plot são as mesmas utilizadas na atividade 2 mínima.

````python
import numpy as np 
import matplotlib.pyplot as plt


theta = np.linspace(0,5* np.pi, 800) #definindo o theta e o intervalo de 0 a 5pi , com 800 subnintervalos

x= 2*np.cos(theta) + np.cos(2*theta)
y= 2*np.sin(theta) - np.sin(2*theta)

plt.plot(x, y, color='blue', label=r'$x = 2\cos(\theta) + \cos(2\theta)$' '\n' r'$y = 2\sin(\theta) - \sin(2\theta)$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Deltoid')
plt.legend()
plt.show()
````
