# Fey's function - atividade 2 avançada

As coordenadas $x$ e $y$ em função de $\theta$ que descrevem a espira de Galileu são dadas por $$x = r\cos{\theta},$$ $$y= r\sin{\theta},$$ onde $$r= e^{\cos{\theta}} - 2\cos{4\theta} + \sin^5{\frac{\theta}{12}}.$$ Dessa forma, foi utilizado um intervalo de $\theta$ conveniente para a visualização da imagem (de 0 a $10\pi$).

Obs: as funções do ```numpy``` e do ```matplotlib.pyplot``` para gerar o plot são as mesmas utilizadas na atividade 2 mínima.

````python
import numpy as np 
import matplotlib.pyplot as plt

theta = np.linspace(0, 10*np.pi,800) #definindo o theta e o intervalo de 0 a 10pi , com 800 subnintervalos

r= np.exp(np.cos(theta)) - 2* np.cos(4*theta) + (np.sin(theta/12))**5
x= r* np.cos(theta)
y= r* np.sin(theta)

plt.plot(x, y, color = 'blue')
plt.title('Fey’s function ')
plt.show()
````
