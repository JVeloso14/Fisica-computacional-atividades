# Função sinc(x) utilizando np.savez e np.load - atividade 2 intermediária

Nesse código, foram salvos apenas dois dados: o intervalo de x e a função $sinc(x)$. Nesse sentido, no arquivo sinc(x).py, esses dois dados foram salvos em um arquivo ```.npz```, por meio do ```np.savez```. Assim, o ```np.load``` leu os dois dados nesse arquivo, o que tornou possível de utilizar as variáveis $x$ e $y$ para o plot do gráfico de $sinc(x)$.

````python
import matplotlib.pyplot as plt
import numpy as np

dados = np.load('Dados de sinc(x).npz')

x= dados['x']
y= dados['funcao']


plt.figure(figsize=(8, 6))
plt.title("Gráfico de y = sen(x)/x")
plt.plot(x, y, label='f(x) = sen(x)/x', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.ylim(-0.5, 1.5)  # Limitando o eixo y para melhor visualização
plt.xlim(-50,50)    # Limitando o eixo x para mostrar o comportamento negativo
plt.show()
````
