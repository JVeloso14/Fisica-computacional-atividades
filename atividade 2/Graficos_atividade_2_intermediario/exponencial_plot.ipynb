# Plot exponencial utilizando np.savez e np.load - atividade 2 intermediária

As funções utilizadas nesse código são as mesmas do "Bhaskara da atividade 2 -intermediária". Os dados $x, y_1, y_2, y_3$ foram salvos no ```exponencial_code.py```, por meio da ```np.savez```, e criou um arquivo ```.npz```. A função ```np.load``` leu os dados salvos nesse arquivo npz. 

````python
import numpy as np
import matplotlib.pyplot as plt

dados= np.load('Dados da exponencial.npz')

x = dados['x']
y1 =dados['y1']
y2 = dados['y2']
y3 = dados['y3']

plt.figure(figsize=(8, 6))
plt.title("Gráfico de y = exp(ax)")
plt.plot(x, y1, label='f(x) = exp(x)', color='blue')
plt.plot(x, y2, label='f(x) = exp(-10x)', color='red')
plt.plot(x, y3, label='f(x) = exp(10x)', color='black')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.ylim(-0.5, 8)  # Limitando o eixo y para melhor visualização
plt.xlim(-4, 2)    # Limitando o eixo x para mostrar o comportamento negativo
plt.show()
````
