# Bhaskara plot utilizada np.savez e np.load - atividade 2 intermediária. 

Nesse código, foi complementado o código de Bhaskara utilizado a função ```np.load``` para ler os dados que foram salvos no arquivo ```Dados de Bhaskara.npz```. Esse arquivo foi criado no ```bhaskara_code.py```, por meio da função ```np.savez``` para salvar os dados passados como parametros. Para idenficar os dados, foi definida cada variável por meio de ``` dados['nome_da_variavel_salva']```. 

Obs: O resto do código foi feito na atividade 2 mínima. Porém, com a adição do ```np.atleast_1d(dados['raizes'])``` garante que tenha pelo menos uma dimensão, evitando assim possíveis problemas (nesse caso temos uma array contendo as raízes).

````python
import numpy as np
import matplotlib.pyplot as plt

dados = np.load("Dados de Bhaskara.npz")

a =dados['a']
b= dados['b']
c= dados['c']
d= dados['d']
e=dados['e']
f=dados['f']
x= dados['x']
y= dados['y']
y2 = dados['y2']
raizes = dados['raizes']
raizes2 = dados['raizes2']


raizes = np.atleast_1d(dados['raizes'])
raizes2 = np.atleast_1d(dados['raizes2'])
# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.title("Gráfico de y = ax² + bx + c")
plt.plot(x, y, label=f'f(x) = {a}x² + {b}x + {c}', color='blue')
plt.plot(x, y2, label=f'f(x) = {d}x² + {e}x + {f}', color='red')


# Destacando as raízes no gráfico
for raiz in raizes:
    plt.scatter(raiz, 0, color='red', zorder=5, label=f'Raiz: x = {raiz:.0f}')  # Marcador das raízes

for raiz in raizes2:
    plt.scatter(raiz, 0, color='red', zorder=5, label=f'Raiz: x = {raiz:.0f}')  # Marcador das raízes

# Configurações do gráfico
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.ylim(-1, 6)  # Limitando o eixo y para melhor visualização
plt.xlim(-1, 6)  # Limitando o eixo x para mostrar o comportamento negativo
plt.show()
````
