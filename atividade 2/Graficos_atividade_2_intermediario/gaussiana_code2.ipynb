# Função Gaussiana utilizando o np.load e np.savez - atividade 2 intermediária

De forma análoga as atividades 2 intermediárias, foi utilizado o ```np.load``` para ler os dados que foram passados como parametros no ```np.savez``` do arquivo gaussiana_code2.py 

Obs: As funções utilizadas nesse código foram as mesmas abordadas nas outras atividades 2 intermediárias.

````python
import numpy as np
import matplotlib.pyplot as plt


dados= np.load('Dados da gaussiana.npz')

x =dados['x']
y1 = dados['y1']
y2 = dados['y2']
y3 = dados['y3']
y4 = dados['y4']


plt.figure(figsize=(8, 6))
plt.title("Gráfico de y = exp(-x^2)")
plt.plot(x, y1, label='f(x) = exp(-x^2)', color='blue')
plt.plot(x, y2, label='f(x) = 2exp(-x^2)', color='red')
plt.plot(x, y3, label='f(x) = exp(-2x^2)', color='black')
plt.plot(x, y4, label='f(x) = exp(-2x^2)', color='green')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.ylim(-0.5, 1.5)  # Limitando o eixo y para melhor visualização
plt.xlim(-10,10)    # Limitando o eixo x para mostrar o comportamento negativo
plt.show()
````
