```python
#Delta de Dirac utilizando séries de Fourier 

import matplotlib.pyplot as plt
import numpy as np 

#L = 4

x= np.linspace(-2 , 2*np.pi , 2000)

def Fourier_delta_Dirac(N):
    soma = 0
    for i in range(1,N-1):
        soma += (1/2*4) + (1/4)*np.cos((i * np.pi *x)/4)
    return soma

soma  = Fourier_delta_Dirac(100000)
plt.plot(x, soma)
plt.show()
```
